```python
from sklearn.feature_extraction.text import TfidfVectorizer

texts = [
    "Patient has mild fever",
    "Patient diagnosed with pneumonia",
    "Patient is healthy",
    "Diabetes risk is high"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(texts)

print(X.toarray())

print(vectorizer.get_feature_names_out())
```

```python
from transformers import pipeline

generator = pipeline(
    task="text-generation",
    model="distilgpt2"
)
```

```python
text = """
Patient Report:
Glucose Level: High
BMI: High
Prediction: High Risk of Diabetes

Generate a simple explanation for the patient.
"""

result = generator(
    text,
    max_new_tokens=50,
    do_sample=False
)

print(result[0]["generated_text"])
```
